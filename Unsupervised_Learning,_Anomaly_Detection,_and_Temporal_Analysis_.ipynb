1. : What is Dimensionality Reduction? Why is it important in machine
learning?

-> Dimensionality Reduction is the process of reducing the number of input features (variables or dimensions) in a dataset while retaining as much important information as possible.


*Avoid Overfitting :	Too many features can cause the model to learn noise instead of actual patterns. Reducing dimensions helps in generalization.

*Improves Training Speed:	Fewer features mean fewer computations, which makes model training and prediction faster.

*Reduces Storage and Memory: Needs	Smaller datasets are easier to store, process, and visualize.

*Better Visualization:	Helps visualize data in 2D or 3D plots for understanding clusters or patterns.

*Removes Multicollinearity:	Eliminates redundant or highly correlated features that can confuse the model.

*Enhances Interpretability:	Simplifies the model, making it easier to understand and explain results.


2.  Name and briefly describe three common dimensionality reduction
techniques.

-> I. Principal Component Analysis (PCA)

Type: Linear technique

Description:
PCA transforms the original features into a new set of uncorrelated variables called principal components.
These components are ordered so that the first few retain most of the variance (information) in the data.
It’s mainly used to reduce dimensionality while preserving as much information as possible.

Example use: Image compression, noise reduction, exploratory data analysis.

II.Linear Discriminant Analysis (LDA)

Type: Supervised linear technique

Description:
LDA finds a linear combination of features that best separates two or more classes.
Unlike PCA (which ignores class labels), LDA uses the class information to maximize the distance between different classes and minimize variation within each class.

Example use: Classification problems such as face recognition or document classification.

III. t-Distributed Stochastic Neighbor Embedding (t-SNE)

Type: Non-linear technique

Description:
t-SNE is mainly used for visualizing high-dimensional data in 2D or 3D space.
It preserves local structures (similar data points stay close) and helps reveal clusters or patterns that are not visible in the raw data.

Example use: Visualizing word embeddings, image datasets, or high-dimensional features.


3.  What is clustering in unsupervised learning? Mention three popular
clustering algorithms.

 -> Clustering is a technique in unsupervised learning that involves grouping similar data points together based on their features — without using any predefined labels.

In other words, it automatically finds patterns or structures in the data by identifying groups (called clusters) where points within the same cluster are more similar to each other than to those in other clusters.

 i. K-Means Clustering

Partitions data into K clusters based on feature similarity. Each cluster has a centroid (center), and data points are assigned to the nearest centroid. It’s simple, efficient, and widely used.

ii. Hierarchical Clustering

Builds a tree-like structure (dendrogram) of clusters by either merging smaller clusters (agglomerative) or splitting larger ones (divisive). It doesn’t need a fixed number of clusters in advance.

iii. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Groups points that are closely packed together and marks points in low-density regions as outliers. It’s great for irregular-shaped clusters and noisy data.

4. Explain the concept of anomaly detection and its significance.

  -> Anomaly detection (also called outlier detection) is the process of identifying data points or patterns that deviate significantly from the expected or normal behavior in a dataset.

In simple terms, it helps find unusual or rare events that don’t fit the general pattern.

Fraud Detection

Identifies abnormal financial transactions or insurance claims.

Fault Detection

Detects early signs of equipment failure in manufacturing or IoT systems.

 Cybersecurity

Flags suspicious network traffic or login behavior.

Healthcare

detects abnormal patient readings or medical results.

Data Quality

Helps find data entry errors or inconsistencies.

5. List and briefly describe three types of anomaly detection techniques.

-> i. Statistical (Model-Based) Methods

Concept:
These methods assume that normal data follows a specific statistical distribution (like Gaussian/normal distribution).
Any data point that significantly deviates from this distribution is marked as an anomaly.

Examples:

Z-score method: Points with Z-scores beyond a threshold (e.g., ±3) are considered outliers.

Boxplot (IQR method): Points outside the interquartile range are flagged.

Use cases: Simple datasets where the data roughly follows a known distribution.

ii. Machine Learning–Based Methods

Concept:
These algorithms learn patterns from data and detect points that don’t fit those patterns.
They can handle complex, high-dimensional datasets.

Examples:

Isolation Forest: Isolates anomalies based on how easily they can be separated from normal data.

One-Class SVM: Learns a boundary around normal data and flags points lying outside it as anomalies.

k-Nearest Neighbors (k-NN): Points far from their nearest neighbors are considered anomalies.

Use cases: Fraud detection, network intrusion, equipment fault detection.

iii. Deep Learning–Based Methods

Concept:
Used for complex and high-dimensional data like images, videos, or sensor signals.
Neural networks learn compact representations of normal data; anomalies are identified when reconstruction or prediction errors are high.

Examples:

Autoencoders: Train to reconstruct normal data; anomalies have higher reconstruction errors.

Recurrent Neural Networks (RNNs): Detect abnormal sequences or time-series patterns.

Use cases: Anomaly detection in IoT systems, cybersecurity, and healthcare monitoring.

6. What is time series analysis? Mention two key components of time series data.

-> Time Series Analysis is a statistical technique used to analyze data points collected or recorded over time — usually at regular intervals (daily, monthly, yearly, etc.).

Two Key Components of Time Series Data

Component	Description

i. Trend	The long-term movement or direction in the data over time. It shows whether the data is increasing, decreasing, or remaining stable.
Example: A steady rise in smartphone sales over several years.

ii. Seasonality	The repeating short-term pattern or fluctuation that occurs at regular intervals (daily, monthly, yearly, etc.) due to seasonal factors.
Example: Higher ice cream sales in summer and lower in winter.


7. Describe the difference between seasonality and cyclic behavior in time series.

-> i. Seasonality

Definition:
Seasonality refers to regular and predictable patterns in a time series that repeat at fixed time intervals, such as every day, month, quarter, or year.

Key Characteristics:

Fixed and known period (e.g., every 12 months, every 7 days)

Caused by seasonal factors like weather, holidays, or business cycles

The pattern repeats consistently over time

Example:

Higher ice cream sales every summer.

Increased electricity consumption every winter.

More online shopping during festive seasons.

ii. Cyclic Behavior

Definition:
Cyclic behavior refers to fluctuations in a time series that occur over irregular or longer intervals, often influenced by economic or business conditions rather than fixed calendar periods.

Key Characteristics:

Not fixed or regular — cycle length varies

Often related to economic expansions and contractions

Duration is longer (can last several years)

Harder to predict than seasonality

Example:

Business cycles showing periods of growth and recession.

Long-term ups and downs in housing or stock markets.


8. Write Python code to perform K-means clustering on a sample dataset.



```python
# Step 1: Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Step 2: Create a sample dataset
# make_blobs generates random data for clustering
X, y = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=1.0)

# Step 3: Visualize the data (before clustering)
plt.scatter(X[:, 0], X[:, 1], s=50, color='gray')
plt.title("Sample Data (Before Clustering)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Step 4: Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Step 5: Get cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Step 6: Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X')  # Cluster centers
plt.title("K-Means Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

9. What is inheritance in OOP? Provide a simple example in Python.

  -> Inheritance is one of the core concepts of OOP.
It allows a class (child or subclass) to inherit attributes and methods from another class (parent or base class).

This promotes code reusability, organization, and extensibility.

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "Some generic sound"

# Child class inherits from Animal
class Dog(Animal):
    def speak(self):  # Overriding parent method
        return "Woof!"

# Another child class
class Cat(Animal):
    def speak(self):
        return "Meow!"

# Create objects
dog1 = Dog("Buddy")
cat1 = Cat("Whiskers")

# Access methods from parent and child
print(dog1.name, "says:", dog1.speak())
print(cat1.name, "says:", cat1.speak())
```

10. How can time series analysis be used for anomaly detection?

   -> Time Series Anomaly Detection involves identifying unusual patterns, sudden spikes, drops, or deviations in data that is recorded over time.

It helps detect events that don’t follow the expected temporal trend or seasonal pattern.

Time series analysis builds a model of normal behavior over time — considering trend, seasonality, and cyclic behavior — and then detects when observed values deviate significantly from the model’s predictions.

Statistical Methods	Use moving averages or standard deviation thresholds (e.g., Z-score > 3).

ARIMA Models	Forecast future values and flag points with large prediction errors.

Machine Learning Methods	Use Isolation Forest, One-Class SVM, or Prophet for pattern detection.

Deep Learning Models	LSTM or Autoencoder models detect anomalies in complex, non-linear time series.
